```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import os

import matplotlib.pyplot as plt
from IPython.display import display
plt.style.use("fivethirtyeight")
sns.set_style({'font.sans-serif':['simhei','Arial']})
%matplotlib inline
```

```python
saved_style_state = mpl.rcParams.copy() #give us a style state to go back to
```

```python
if os.path.isfile("StudentsPerformance.csv"):
    filepath = "StudentsPerformance.csv"
    print("loading from file")
else:
    filepath = "https://www.kaggle.com/spscientist/students-performance-in-exams/downloads/students-performance-in-exams.zip/1"
    print("loading from the internet")

StudentsPerformance = pd.read_csv(filepath)
print("done")
```

```python
sns.set()
sns.set_style("ticks")
plt.axis('equal')
StudentsPerformance['gender'].value_counts().plot.pie(autopct='%1.2f%%')
```

```python
fig, ax=plt.subplots(1,3,figsize=(15,4))
StudentsPerformance.boxplot(column='math score', by='gender', showfliers=False, ax=ax[0])
StudentsPerformance.boxplot(column='reading score', by='gender', showfliers=False, ax=ax[1])
StudentsPerformance.boxplot(column='writing score', by='gender', showfliers=False, ax=ax[2])

```

```python
plt.figure(figsize=(20,20))
sns.catplot(x="gender", y="writing score",
                 hue="parental level of education",
                 data=StudentsPerformance, kind="bar")
plt.title('parental level of education gender & writing score')
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.catplot(x="gender", y="math score",
                 hue="lunch",
                 data=StudentsPerformance, kind="bar")
plt.title('lunch gender & math score')
plt.show() 
plt.figure(figsize=(20,20))
sns.catplot(x="gender", y="reading score",
                 hue="lunch",
                 data=StudentsPerformance, kind="bar")
plt.title('lunch gender & reading score')
plt.show()
plt.figure(figsize=(20,20))
sns.catplot(x="gender", y="writing score",
                 hue="lunch",
                 data=StudentsPerformance, kind="bar")
plt.title('lunch gender & writing score')
plt.show()
```

```python

plt.figure(figsize=(20,20))
sns.catplot(x="gender", y="math score",
                 hue="test preparation course",
                 data=StudentsPerformance, kind="bar")
plt.title('test preparation course gender & math score')
plt.show() 
plt.figure(figsize=(20,20))
sns.catplot(x="gender", y="reading score",
                 hue="test preparation course",
                 data=StudentsPerformance, kind="bar")
plt.title('test preparation course gender & reading score')
plt.show()
plt.figure(figsize=(20,20))
sns.catplot(x="gender", y="writing score",
                 hue="test preparation course",
                 data=StudentsPerformance, kind="bar")
plt.title('test preparation course gender & writing score')
plt.show()
```

```python
sns.set()
sns.set_style("ticks")
plt.axis('equal')
StudentsPerformance['race/ethnicity'].value_counts().plot.pie(autopct='%1.2f%%')
```

```python
StudentsPerformance_count = StudentsPerformance.groupby('race/ethnicity')['math score'].count().sort_values(ascending=False).to_frame().reset_index()
StudentsPerformance_mean = StudentsPerformance.groupby('race/ethnicity')['math score'].mean().sort_values(ascending=False).to_frame().reset_index()

f, [ax1,ax2,ax3] = plt.subplots(3,1,figsize=(20,15))
sns.barplot(x='race/ethnicity', y='math score', palette="Blues_d", data=StudentsPerformance_mean, ax=ax1)
ax1.set_title('math score',fontsize=15)
ax1.set_xlabel('race/ethinicity')
ax1.set_ylabel('average score')

sns.barplot(x='race/ethnicity', y='math score', palette="Greens_d", data=StudentsPerformance_count, ax=ax2)
ax2.set_title('total amount',fontsize=15)
ax2.set_xlabel('race/ethnicity')
ax2.set_ylabel('counts')

sns.boxplot(x='race/ethnicity', y='math score', data=StudentsPerformance, ax=ax3)
ax3.set_title('total score',fontsize=15)
ax3.set_xlabel('race/ethnicity')
ax3.set_ylabel('total score')

plt.show()
```

```python
StudentsPerformance_count = StudentsPerformance.groupby('race/ethnicity')['reading score'].count().sort_values(ascending=False).to_frame().reset_index()
StudentsPerformance_mean = StudentsPerformance.groupby('race/ethnicity')['reading score'].mean().sort_values(ascending=False).to_frame().reset_index()

f, [ax1,ax2,ax3] = plt.subplots(3,1,figsize=(20,15))
sns.barplot(x='race/ethnicity', y='reading score', palette="Blues_d", data=StudentsPerformance_mean, ax=ax1)
ax1.set_title('reading score',fontsize=15)
ax1.set_xlabel('race/ethinicity')
ax1.set_ylabel('average score')

sns.barplot(x='race/ethnicity', y='reading score', palette="Greens_d", data=StudentsPerformance_count, ax=ax2)
ax2.set_title('total amount',fontsize=15)
ax2.set_xlabel('race/ethnicity')
ax2.set_ylabel('counts')

sns.boxplot(x='race/ethnicity', y='reading score', data=StudentsPerformance, ax=ax3)
ax3.set_title('total score',fontsize=15)
ax3.set_xlabel('race/ethnicity')
ax3.set_ylabel('total score')

plt.show()
```

```python
StudentsPerformance_count = StudentsPerformance.groupby('race/ethnicity')['writing score'].count().sort_values(ascending=False).to_frame().reset_index()
StudentsPerformance_mean = StudentsPerformance.groupby('race/ethnicity')['writing score'].mean().sort_values(ascending=False).to_frame().reset_index()

f, [ax1,ax2,ax3] = plt.subplots(3,1,figsize=(20,15))
sns.barplot(x='race/ethnicity', y='writing score', palette="Blues_d", data=StudentsPerformance_mean, ax=ax1)
ax1.set_title('writing score',fontsize=15)
ax1.set_xlabel('race/ethinicity')
ax1.set_ylabel('average score')

sns.barplot(x='race/ethnicity', y='writing score', palette="Greens_d", data=StudentsPerformance_count, ax=ax2)
ax2.set_title('total amount',fontsize=15)
ax2.set_xlabel('race/ethnicity')
ax2.set_ylabel('counts')

sns.boxplot(x='race/ethnicity', y='writing score', data=StudentsPerformance, ax=ax3)
ax3.set_title('total score',fontsize=15)
ax3.set_xlabel('race/ethnicity')
ax3.set_ylabel('total score')

plt.show()
```
